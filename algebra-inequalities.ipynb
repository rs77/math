# Inequalities

In this section we will learn how to solve inequality equations where there is an unknown variable in the denominator.

For example, solve...

$\frac{x^2 + 5}{x} \gt 6$

### Multiply by $x^2$

To answer this equation we need to multiply both sides by $x^2$, why?

As $x$ can represent any number we know when dealing with inequalities that if we **multiply or divide by a negative number** that the inequality sign needs to be **reversed**.

To remove any doubt that we are handling a negative number, we instead multiply by an expression that will **always** be positive: $x^2$

Multiplying by this means that we don't have to *worry* about switching the inequality sign.

### Answer

Answering the above question then, would produce the following results:

$$\begin{align*}
\frac{x^2 + 5}{x} &\gt 6 && \text{original question} \\
\frac{x^2 + 5}{\cancel{x}} \times x^{\cancel{2} 1} &\gt 6 \times x^2 && \text{multiply both sides by $x^2$} \\
x(x^2 + 5) &\gt 6x^2 && \text{clean up} \\ 
x^3 + 5x &\gt 6x^2 && \text{expand} \\
x^3 + 5x - 6x^2 &\gt \cancel{6x^2 - 6x^2} && \text{subtract $6x^2$ both sides} \\
x(x^2 - 6x + 5) &\gt 0 && \text{factorise $x$ from each term} \\
x(x - 5)(x - 1) &\gt 0 && \text{factorise $x^2 - 6x + 5$} \\
\end{align*}$$

Now that we have our three solutions, we can proceed to plot these on a graph where we know there will be three x-intercepts: $(0, 0), (5, 0), (1, 0)$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

# Define the function
def f(x):
    return x**3 - 6*x**2 + 5*x

# Create x values
x = np.linspace(-2, 5, 400)

# Calculate y values
y = f(x)

# Find x-intercepts using root finding
x_intercepts = []
for guess in [0, 1, 4]:  # initial guesses based on approximate root locations
    root = optimize.fsolve(f, guess)[0]
    # Only add unique roots within our x range
    if -2 <= root <= 5 and round(root, 4) not in [round(r, 4) for r in x_intercepts]:
        x_intercepts.append(root)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label='f(x) = x³ - 6x² + 5x')

# Plot x and y axes
plt.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
plt.axvline(x=0, color='k', linestyle='-', linewidth=0.5)

# Plot and label x-intercepts
for x_int in x_intercepts:
    plt.plot(x_int, 0, 'ro')  # red dots at intercepts
    plt.annotate(f'({x_int:.2f}, 0)', 
                 xy=(x_int, 0), 
                 xytext=(5, 5), 
                 textcoords='offset points')

# Customize the plot
plt.grid(True, linestyle='--', alpha=0.7)
plt.title('Graph of f(x) = x³ - 6x² + 5x with X-intercepts')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Show the plot
plt.show()

# Print the x-intercepts
print("X-intercepts:", [f"{x:.4f}" for x in x_intercepts])
```
