# Present Value

The concept of present value is to calculate the value of payments made in the future, or the amount needed now for consisently making withdrawals in the future.

What would you prefer:

$\$100$ today or $\$100$ in a year's time?

Obviously you'd prefer the $\$100$ now because with it you have the capacity to invest, say with the bank, and get a return on that money meaning by the end of the year you could have a little more, say $\$105$ because of the interest earned compared to the offer of $\$100$ in a year's time.

That's the time value of money.

If we are receiving payments in the future we can calculate what the equivalent present value of those future payments will be.

Continuing with our simple example above, if you were given the offer of $\$100$ in a year's time or $\$95$ now, which would you prefer?

We can answer this question in one of two ways:

1. If we know the bank is offering $5\%$ p.a. return on our money this means that if we take the offer of $\$95$ now this means in one year we will have the following balance:

$$\$95 \times (1 + 0.05)^1 = \$99.75$$

We can see from this simple calculation that we would have **less** than the other offer of $\$100$ and would be slightly better off with the $\$100$ in one year's time.

2. The other way we can calculate this deal is by using the same type of approach but instead of **multiplying** the starting amount, we instead **divide** (or discount) the interest from our future value of $\$100$, it would look like this:

$$\$100 \div (1 + 0.05)^1 = \$95.24$$

This means if we are offered $\$95.24$ **or more now** that we would be better off than the $\$100$ in one year's time.

We can therefore use the principle of **discounting** future payments to calculate what the value of future payments is today.

## Formula

Knowing the discounting principle can then help us to create a formula so that we can easily calculate the present value.

If we use a simple example of receiving future payments every year for the next 3 years in the amount of $\$100$ and the interest rate is $5.0\%$ per year, to calculate the present value of these future payments we would manually calculate each payment as follows:

| Time          |                         Calculation |
|:--------------|------------------------------------:|
| End of year 1 | $\$100 \div (1 + 0.05)^1 = \$95.24$ |
| End of year 2 | $\$100 \div (1 + 0.05)^2 = \$90.70$ |
| End of year 3 | $\$100 \div (1 + 0.05)^3 = \$86.26$ |

Here we have discounted each of the future payments to the present value, if we add all of these values up we will have the total present value of each $\$100$ future payment:

$$\$100 \div (1 + 0.05)^1 + \$100 \div (1 + 0.05)^2 + \$100 \div (1 + 0.05)^3$$
$$\$95.24 + \$90.70 + \$86.26 = \$272.20$$

With this simple calculation we can then translate what we've done into a formula.

Let's label the future payments as $P$, interest rate as $r$ and the number of periods as $n$.

Replacing the numbers with these labels we would have something like this:

$$P \div (1 + r)^1 + P \div (1 + r)^2 + P \div (1 + r)^3$$

If we factorise out $P$ from our formula we get:

$$P \times \Bigg(\frac{1}{(1 + r)^1} + \frac{1}{(1 + r)^2} + \frac{1}{(1 + r)^3}\Bigg)$$

Instead of writing fractions we can replace the powers with a negative sign:

$$P \times \Big((1 + r)^{-1} + (1 + r)^{-2} + (1 + r)^{-3}\Big)$$

### Geometric Series

You should see within the bracketed expression a geometric series: $(1 + r)^{-1} + (1 + r)^{-2} + (1 + r)^{-3}$.

A geometric series has the form: $a + ar + ar^2 + ar^3 + ... + ar^{n-1}$

Where $a$ is the first term and $r$ is the common ratio.

If we compare this to our formula we would have $a = (1 + r)^{-1}$ and the ratio $q = (1 + r)^{-1}$.

This ratio $q$ is calculated by dividing the second term by the first term (or any term followed by the previous term):

$$q = \frac{(1 + r)^{-2}}{(1 + r)^{-1}} = (1 + r)^{-1}$$

So in geometric form we would have $\text{Sum} = a(1 + q + q^2 + q^3 + ... + q^{n-1})$

#### Geometric Series Formula

For any finite series we can use the geometric series formula (found on your [formula sheet](https://www.nsw.gov.au/sites/default/files/noindex/2025-05/mathematics-advanced-ext1-ext2-reference-sheet-nov2019.pdf)):

$$S_{n} = \frac{a(1 - r^{n})}{1 - r} = \frac{a(r^n - 1)}{r - 1}, r \neq 1$$

Substituting our values into this formula would look as follows:

$$S_{n} = \frac{(1 + r)^{-1} \times (1 - ((1 + r)^{-1})^{n})}{1 - (1 + r)^{-1}}$$

If we focus on the denominator first and simplify it...

$$1 - (1 + r)^{-1} = 1 - \frac{1}{1 + r} = \frac{(1 + r) - 1}{1 + r} = \frac{r}{1 + r}$$

And then focus on the numerator next and simplify it...

$$\begin{align*}
(1 + r)^{-1} \times (1 - ((1 + r)^{-1})^{n}) &= \frac{1}{(1 + r)^{1}} \times \Bigg(1 - \frac{1}{(1 + r)^{n}}\Bigg) \\
&= \frac{1}{(1 + r)^{1}} - \frac{1}{(1 + r)^{n}} \\
\end{align*}$$

Now let's combine the numerator and denominator together and simplify:

$$\begin{align*}
\Bigg(\frac{1}{(1 + r)^{1}} - \frac{1}{(1 + r)^{n}}\Bigg) \div \frac{r}{1 + r} \\ \Bigg(\frac{1}{(1 + r)^{1}} - \frac{1}{(1 + r)^{n}}\Bigg) \times \frac{1 + r}{r} && &\text{flip and multiply trailing fraction}
\end{align*}$$

## Applications

Applications where we see present value calculations used are popular where investments are used to help calculate retirement.

For example, how much do I need in my superannuation account so that when I retire I will receive $\$60,000$ per year.

[ASFA even helps provide this information to give you an indication of how much you will need in retirement](https://www.superannuation.asn.au/consumers/retirement-standard/)

Other areas where this calculation is used is in cashing out investments like term deposits or selling a business. You can use the present value formula to determine what your business is worth based on the potential of future payments.

The above values in the table are calculated using the formula:

$$\frac{1 - (1+r)^{-n}}{r}$$

Where $n$ represents the number of periods of the **compounding**. For example, if the interest rate is *per annum* but the compounding is done *monthly* then $n$ would represent the number of periods of compounding in the year, being $12$.

This means $r$ represents the interest rate **at the compouning frequency**. Again if the interest rate is quoted as *p.a. (per annum)* but the compounding is done *monthly* then $r$ would need to be divided by $12$ to obtain the interest rate at the compounding frequency.

To use the above formula and to compare this to the table of values, if we choose an interest rate of $1.0\% \ \text{p.a.}$ with a $20$ year term this would mean $n=20$ and inserting the values into the formula would produce:

$$\frac{1 - (1 + 0.01)^{-20}}{0.01} = 18.04555297 \approx 18.046$$
